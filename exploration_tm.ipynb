```python
import pandas as pd

turnstiles = pd.read_pickle('final_turnstile_data.pkl')
```

```python
turnstiles.head()
```

```python
turnstiles.head()
turnstiles.describe()
```

```python
turnstiles.loc[turnstiles['entries_diff'] < 0, 'entries_diff'] = None
turnstiles.loc[turnstiles['exit_diff'] < 0, 'exit_diff'] = None
```

```python
turnstiles['day'] = [datetime.datetime.strftime(x, '%Y-%m-%d') for x in turnstiles['datetime_clean']]
turnstiles['all_traffic'] = turnstiles['entries_diff'] + turnstiles['exit_diff']
```

```python
turnstiles.head()
```

```python
# See if there are multiple stations per line or division
turnstiles.groupby(['station', 'linename', 'division']).sum()
```

```python
stations_day = turnstiles.groupby(['station', 'day']).sum()
stations_day.reset_index(inplace=True)
```

```python
stations_day.head()
```

```python
daily_mean = stations_day.groupby('station')['all_traffic'].mean()
daily_mean.head()
```
